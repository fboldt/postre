```python
from sklearn.datasets import make_swiss_roll

X, t = make_swiss_roll(n_samples=1000, noise=0.2, random_state=42)
X.shape
```

```python
import matplotlib.pyplot as plt

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.view_init(10, -70)

ax.scatter(X[:,0], X[:,1], X[:,2], c=t)
plt.show()
```

```python
from sklearn.decomposition import PCA

pca = PCA(n_components=2)
X2D = pca.fit_transform(X)
plt.scatter(X2D[:,0], X2D[:,1], c=t, alpha=0.3)
plt.show()
```

```python
from sklearn.decomposition import KernelPCA

linpca = KernelPCA(kernel='linear', n_components=2)
X2D = linpca.fit_transform(X)
plt.scatter(X2D[:,0], X2D[:,1], c=t, alpha=0.3)
plt.show()
```

```python
rbfpca = KernelPCA(kernel='rbf', gamma=0.002, n_components=2)
X2D = rbfpca.fit_transform(X)
plt.scatter(X2D[:,0], X2D[:,1], c=t, alpha=0.3)
plt.show()
```

```python
sigpca = KernelPCA(kernel='sigmoid', gamma=0.002, n_components=2)
X2D = sigpca.fit_transform(X)
plt.scatter(X2D[:,0], X2D[:,1], c=t, alpha=0.3)
plt.show()
```
