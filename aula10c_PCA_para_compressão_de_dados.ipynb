```python
from sklearn.datasets import fetch_openml

mnist = fetch_openml('mnist_784', as_frame=False, parser="auto")
X_tr, y_tr = mnist.data[:60_000], mnist.target[:60_000]
X_te, y_te = mnist.data[60_000:], mnist.target[60_000:]
```

```python
X_tr.shape
```

```python
import numpy as np
np.sqrt(784)
```

```python
import matplotlib.pyplot as plt

k=50
plt.figure(figsize=(8,2))
for index, X in enumerate(X_tr[:50]):
  plt.subplot(k//10, 10, index+1)
  plt.imshow(X.reshape(28,28), cmap='binary', interpolation="bilinear")
  plt.axis("off")
plt.show()
```

```python
from sklearn.decomposition import PCA

pca = PCA(0.95)
X_reduced = pca.fit_transform(X_tr, y_tr)
X_reduced.shape
```

```python
154/784
```

```python
X_rec = pca.inverse_transform(X_reduced)
X_rec.shape
```

```python
k=50
plt.figure(figsize=(8,2))
for index, X in enumerate(X_rec[:k]):
  plt.subplot(k//10, 10, index+1)
  plt.imshow(X.reshape(28,28), cmap='binary', interpolation="bilinear")
  plt.axis("off")
plt.show()
plt.figure(figsize=(8,2))
for index, X in enumerate(X_tr[:k]):
  plt.subplot(k//10, 10, index+1)
  plt.imshow(X.reshape(28,28), cmap='binary', interpolation="bilinear")
  plt.axis("off")
plt.show()
```
